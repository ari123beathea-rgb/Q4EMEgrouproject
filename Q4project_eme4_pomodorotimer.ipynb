```python
import time

def start_countdown(duration, workorbreak):
    print(f"\n--- {workorbreak}: Timer ({duration} seconds) ---")
    while duration > 0:
        mins, secs = divmod(duration, 60)
        timer_output = f'{int(mins):02d}:{int(secs):02d}'
        print(f"\r{workorbreak}: {timer_output}", end='')
        time.sleep(1)
        duration -= 1
    print(f"\r{workorbreak}: 00:00 - {workorbreak} interval finished!")

def pomodorotimer(overalltime, workduration, breakduration):
    sessionstarttime = time.time()
    sessionendtime = sessionstarttime + overalltime
    pomodorocount = 0

    while time.time() < sessionendtime:
        pomodorocount += 1
        print(f"--- Pomodoro Cycle #{pomodorocount} ---")

        remainingsessiontime = sessionendtime - time.time()
        current_work_duration = min(workduration, remainingsessiontime)
        if current_work_duration <= 0:
            print("Not enough time left for a full work interval. Session over. ")
            break
        start_countdown(int(current_work_duration), "Work")

        if time.time() >= sessionendtime:
            print("Total session time reached. Congrats!")
            break

        remainingsession_time = sessionendtime - time.time()
        current_break_duration = min(breakduration, remainingsessiontime)
        if current_break_duration <= 0:
            print("Not enough time left for a full break interval. Session ending.")
            break
        start_countdown(int(current_break_duration), "Break")

    print("Pomodoro session completed!")


print ("Good day!")
overalltime = float(input("How much time do you want to use? (in seconds) "))
workduration = int(input("How long do you want to work per interval? "))
breakduration = int(input("How long do you want your breaks to be? "))

started = ""
while started.lower() != "yes":
  started = input("Do you wish to start the timer? (yes/no) ")
if started.lower() == "yes":
  pomodorotimer(overalltime, workduration, breakduration)
else:
    print("Timer not started.")
```
